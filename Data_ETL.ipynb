### Create new data with select columns

```python
#create new column with just the make of the car
new_cars_df['Make'] = new_cars_df['Model'].str.split(' ').str[1]
new_cars_df.head()
```

```python
#create new column with just the model of the car
new_cars_df['Model_Specific'] = new_cars_df['Model'].str.split(' ').str[2]
new_cars_df.head()
```

```python
#create new column with just the Year of the car
new_cars_df['Model_Year'] = new_cars_df['Model'].str.split(' ').str[0]
new_cars_df.head()
```
